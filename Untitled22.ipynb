```python
import pandas as pd
import numpy as np
from sklearn.feature_selection import VarianceThreshold, SelectKBest, f_classif
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE
from google.colab import drive
drive.mount('/content/drive')
```

This appears to be a gene expression microarray dataset for leukemia classification. Let me examine the key characteristics:

7129 columns: 7128 gene expression features (Gene_1 to Gene_7128) + 1 target variable (Leukemia_class)

Separator: Semicolon (;)

Data type: Numerical gene expression values (likely normalized or transformed)

Target: Binary classification (acute lymphoblastic leukemia [ALL = 0] vs. acute myeloid leukemia [AML = 1])

```python
df =pd.read_csv("/content/drive/MyDrive/leukemiamicroarray.csv", sep=';')
df
```

```python
print(f"Original data shape: {df.shape}")
print(f"Features: {df.shape[1]-1}, Samples: {df.shape[0]}")
print(f"Class distribution: {df['Leukemia_class'].value_counts().to_dict()}")
print(f"p/n ratio: {(df.shape[1]-1)/df.shape[0]:.1f}:1")

X = df.drop('Leukemia_class', axis=1)
y = df['Leukemia_class']
```

```python
# Remove genes that don't vary much between samples
var_threshold = VarianceThreshold(threshold=0.05)
X_filtered = var_threshold.fit_transform(X)

print(f"Original features: {X.shape[1]}")
print(f"After variance filtering: {X_filtered.shape[1]}")
print(f"Removed {X.shape[1] - X_filtered.shape[1]} low-variance genes")

# Get the remaining feature names
remaining_features = X.columns[var_threshold.get_support()]
print(f"New p/n ratio: {X_filtered.shape[1]}/72 = {X_filtered.shape[1]/72:.1f}:1")


```

```python
# Use the same conservative approach: select top 15 genes
feature_selector = SelectKBest(f_classif, k=15)
X_selected = feature_selector.fit_transform(X_filtered, y)

print(f"After statistical selection: {X_selected.shape[1]} features")
print(f"Final p/n ratio: {X_selected.shape[1]}/72 = {X_selected.shape[1]/72:.1f}:1")

# Get the selected gene names
selected_genes = remaining_features[feature_selector.get_support()]
print(f"✓ Selected {len(selected_genes)} most statistically significant genes")
```

```python
print(f"Before balancing:")
print(f"  Class 0: {sum(y==0)} samples")
print(f"  Class 1: {sum(y==1)} samples")

# Apply SMOTE to balance classes
smote = SMOTE(random_state=42)
X_balanced, y_balanced = smote.fit_resample(X_selected, y)

print(f"After balancing:")
print(f"  Class 0: {sum(y_balanced==0)} samples")
print(f"  Class 1: {sum(y_balanced==1)} samples")
print(f"Balanced data shape: {X_balanced.shape}")
```

```python
# Split the balanced data
X_train, X_test, y_train, y_test = train_test_split(
    X_balanced, y_balanced,
    test_size=0.25,
    random_state=42,
    stratify=y_balanced
)

print(f"Training set: {X_train.shape}")
print(f"Test set: {X_test.shape}")
print(f"Training classes: {np.unique(y_train, return_counts=True)}")
print(f"Test classes: {np.unique(y_test, return_counts=True)}")
```

```python
# Convert back to DataFrames with proper column names
X_train_df = pd.DataFrame(X_train, columns=selected_genes)
X_test_df = pd.DataFrame(X_test, columns=selected_genes)

# Add target columns
train_data = X_train_df.copy()
train_data['Leukemia_class'] = y_train

test_data = X_test_df.copy()
test_data['Leukemia_class'] = y_test

print(f"  - leukemia_preprocessed_train.csv ({train_data.shape})")
print(f"  - leukemia_preprocessed_test.csv ({test_data.shape})")
```

FINAL PREPROCESSING SUMMARY:
* Started with: 7129 genes, 72 samples
* Final training: 15 genes, 73 samples
* Final testing: 15 genes, 25 samples
* Class balance: Perfectly balanced
* p/n ratio improved: 99.0:1 → 0.2:1

# Plotting

```python
# Create subplots
fig, axes = plt.subplots(2, 3, figsize=(18, 12))

# ===== BEFORE PREPROCESSING =====

# 1. Original class distribution
axes[0,0].bar(['Class 0', 'Class 1'], [sum(y==0), sum(y==1)], color=['blue', 'red'])
axes[0,0].set_title('BEFORE: Class Distribution\n(49 vs 23 - Imbalanced)')
axes[0,0].set_ylabel('Number of Samples')

# 2. Original feature variance distribution
original_variances = X.var()
axes[0,1].hist(original_variances, bins=50, alpha=0.7, color='green', edgecolor='black')
axes[0,1].axvline(0.05, color='red', linestyle='--', label='Variance threshold (0.05)')
axes[0,1].set_title('BEFORE: Feature Variances\n(7129 genes)')
axes[0,1].set_xlabel('Variance')
axes[0,1].set_ylabel('Number of Genes')
axes[0,1].legend()

# 3. PCA of original data (first 2 components)
pca_original = PCA(n_components=2)
X_pca_original = pca_original.fit_transform(X)
scatter_original = axes[0,2].scatter(X_pca_original[:, 0], X_pca_original[:, 1], c=y, cmap='coolwarm', alpha=0.7)
axes[0,2].set_title('BEFORE: PCA Plot\n(7129 genes)')
axes[0,2].set_xlabel(f'PC1 ({pca_original.explained_variance_ratio_[0]:.1%})')
axes[0,2].set_ylabel(f'PC2 ({pca_original.explained_variance_ratio_[1]:.1%})')
axes[0,2].legend(handles=scatter_original.legend_elements()[0], labels=['ALL', 'AML'])

# ===== AFTER PREPROCESSING =====

# 1. Balanced class distribution
axes[1,0].bar(['Class 0', 'Class 1'], [sum(y_balanced==0), sum(y_balanced==1)], color=['blue', 'red'])
axes[1,0].set_title('AFTER: Class Distribution\n(49 vs 49 - Balanced)')
axes[1,0].set_ylabel('Number of Samples')

# 2. Selected features variance
selected_variances = X_train_df.var()
axes[1,1].hist(selected_variances, bins=15, alpha=0.7, color='orange', edgecolor='black')
axes[1,1].set_title('AFTER: Selected Features Variance\n(15 genes)')
axes[1,1].set_xlabel('Variance')
axes[1,1].set_ylabel('Number of Genes')

# 3. PCA of preprocessed data
pca_processed = PCA(n_components=2)
X_pca_processed = pca_processed.fit_transform(X_train_df)
scatter_processed = axes[1,2].scatter(X_pca_processed[:, 0], X_pca_processed[:, 1], c=y_train, cmap='coolwarm', alpha=0.7)
axes[1,2].set_title('AFTER: PCA Plot\n(15 genes)')
axes[1,2].set_xlabel(f'PC1 ({pca_processed.explained_variance_ratio_[0]:.1%})')
axes[1,2].set_ylabel(f'PC2 ({pca_processed.explained_variance_ratio_[1]:.1%})')
axes[1,2].legend(handles=scatter_processed.legend_elements()[0], labels=['ALL', 'AML'])

plt.tight_layout()
plt.show()
```
